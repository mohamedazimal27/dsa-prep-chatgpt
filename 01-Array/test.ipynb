2Ô∏è‚É£ Given `arr = [2, 4, 6, 8, 10]`, print all **even numbers**.

```python
arr = [2, 4, 6, 8, 10]

for i, v in enumerate(arr):
    if v % 2 == 0:
        print(v)
```

3Ô∏è‚É£ Given `arr = [3, 1, 7, 9, 2]`, find the **maximum element** manually (without `max()`).

```python
arr = [9,1,7,15,17]
max = arr[0]

for i,v in enumerate(arr):
    if max <= v:
        max = v 
print(max)   
```

4Ô∏è‚É£ Replace every number greater than 5 in a list with `"High"`.
Example: `[4, 7, 2, 9]` ‚Üí `[4, "High", 2, "High"]`


```python
arr = [5,7,2,5.1]

for i,v in enumerate(arr):
    if v > 5:
        arr[i] = "High"
print(arr)
```

5Ô∏è‚É£ Reverse a list **without using** `.reverse()` or slicing `[::-1]`.

```python
arr = [12, 124, 355, 1256, 123]
n = len(arr)

for i in range(n // 2):
    arr[i], arr[n - i - 1] = arr[n - i - 1], arr[i]

print(arr)
```

6Ô∏è‚É£ Print both **index and value** of each element in `arr = [10, 20, 30]` using `enumerate()`.

```python
arr = [10,20,30]

for i,v in enumerate(arr):
    print(f"{i} {v}")
```

7Ô∏è‚É£ Given `nums = [1, 2, 3, 4, 5]`, create a **new list with each number squared** using a **list comprehension**.

```python
nums = [1, 2, 3, 4, 5]
squared = [num*num for num in nums]

print(nums)
print(squared)
```

8Ô∏è‚É£ Filter out all **negative numbers** from `arr = [5, -3, 2, -1, 0]` using list comprehension.

```python
arr = [5, -3, 2, -1, 0]
remove_negative = [num for num in arr if num >= 0]
print(remove_negative)
```

9Ô∏è‚É£ Given `matrix = [[1,2,3], [4,5,6]]`, print every element using **nested loops**.

```python
matrix = [[1,2,3], [4,5,6]]
for row in matrix:
    for val in row:
        print(val)
```
