write a short Python snippet to build the prefix sum list for: arr = [5, 2, 7, 3]


```python
arr = [5, 3, 7, 2]
prefix_sum_arr = [arr[0]]
for i in range(1, len(arr)):
    prefix_sum_arr.append(prefix_sum_arr[i-1]+arr[i]) 

print(prefix_sum_arr)
```
