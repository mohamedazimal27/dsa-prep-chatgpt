````python
def subarray_sum(arr, L, R):
    prefix_arr = [arr[0]]
    for i in range(1,len(arr)):
        prefix_arr.append(prefix_arr[i-1] + arr[i])
    
    if L == 0:
        return prefix_arr[R]
    else:
        return prefix_arr[R] - prefix_arr[L-1]
    
arr = [5, 2, 7, 3]
print(subarray_sum(arr, 1, 3))  # Expected 12 (2 + 7 + 3)
print(subarray_sum(arr, 0, 2))  # Expected 14 (5 + 2 + 7)
print(subarray_sum(arr, 2, 2))  # Expected 7
````
